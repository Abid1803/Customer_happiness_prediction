I am using Random forest scores to determine the important features from the datset*

```python
"""
why my accuracy is still 50% despite appling different model..(logistic, random forest, xgboost, svc, linearsvc)
1. looking for important feaures in the dataset again through random forest scores
"""


path="/content/drive/My Drive/26_09_2025/customer_dataset.csv"
import pandas as pd
df=pd.read_csv(path)



import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

numerical_features = ['Plan_Duration_Months', 'Monthly_Charge_USD', 'Data_Usage_GB']
categorical_features = ['Sex', 'City']

X = df[numerical_features + categorical_features]
y = df['Is_Happy']

X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

feature_importances = pd.DataFrame({
    'feature': X_encoded.columns,
    'importance': model.feature_importances_
}).sort_values('importance', ascending=False)

print("\nFeature Importance Scores:")
print(feature_importances)

```

*2.dropped the irrelevent feaures or features that donot contribute in the prediction of target variable("Is_Happy").
now among those imp feature we are look for the correlation
if the correlation is near zero.. that means that particulate feature's change in value doesnt impact the target variable...(Weak dataset)*

```python
"""
2. as the important feaures are Monthly_Charge_USD    ,Data_Usage_GB    ,Plan_Duration_Months     and none other is contributing to the prediction...(came to this conclusion via
last snippets feature importance)
so finding the correlation between those to target vairable)"""

import pandas as pd

correlation_matrix=df.corr(numeric_only=True)

target_correlation=correlation_matrix['Is_Happy'].sort_values(ascending=False)

print("Correlation with 'Is_Happy' Target Variable:")
print(target_correlation)
```
